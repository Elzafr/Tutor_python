```python
bangundatar1 = 'persegipanjang1'
Panjang1 = 50
lebar1 = 20

bangundatar2 = 'persegipanjang2'
Panjang2 = 60
lebar2 = 10

bangundatar3 = 'persegipanjang3'
Panjang3 = 40
lebar3 = 30
```

```python
def PersegiPanjang(bangundatar, panjang, lebar):
    Luas = panjang * lebar
    print('Luas Persegi Panjang = ', Luas)
    if Luas>700:
        return bangundatar + ' Luas/'
    else:
        return bangundatar + ' Tidak Luas/'
```

```python
Luas1 = PersegiPanjang(bangundatar1, Panjang1, lebar1)
Luas2 = PersegiPanjang(bangundatar2, Panjang2, lebar2)
Luas3 = PersegiPanjang(bangundatar3, Panjang3, lebar3)

print (Luas1, Luas2, Luas3)
```
