```python
import pandas as pd
```

```python
data = {'fruit': ['STRAWBERRY', 'LEMON', 'Pancake'],
       'grams': [50,80,100],
       'price': [1, 2, 8]}

df = pd.DataFrame(data)
df
```

```python
df['fruit'] = df['fruit'].str.lower()
```

```python
df
```

```python
fruit_mapper = {'strawberry':'berry', 'lemon':'citrus', 'pancake':'not_fruit'}
df['type'] = df['fruit'].apply(lambda x: fruit_mapper.get(x))
df
```

```python
df['price/100grams'] = df['grams']/100 * df['price']
```
